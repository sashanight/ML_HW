** Задача 3 **
Докажите, что the k-means всегда сходится.

Ответ: k-means пытается минимизировать сумму квадратов расстояний между каждой точкой $x_i$ и центром кластера $y_j$, к которому он относится. Эта целевая функция k-means будет уменьшаться всякий раз, когда есть изменения в присвоении или шагах измерения, и сближение  (сходимость в одной точке) гарантировано за конечное число итераций. Почему она уменьшается вполне понятно: если точка осталась в старом кластере, то это происходит в силу пересчета центра(он теперь в центре масс), если точка перешла в новый кластер, то уменьшение в силу того что новый центр оказался ближе старого.

Для сжатия размерности пространства алгоритм PCA применяется датасету с количеством признаков $D = 100$. Наблюдается следующий спектр собственных значений матрицы объектов-признаков.

Ответье на вопросы

* 1) Высокая ли эффективная размерность пространства признаков (intrinsic dimensionality) (насколько она близка к 100)?
* 2) Можно ли перевести датасет с помощью PCA в пространство меньшей размерности с минимальными потерями точности? Если да, то чему примерно будет равна размернось 

Ответ: E(m) = $\frac{\lambda_{m+1} + ... + \lambda_n}{\lambda_1 + ... + \lambda_n }$ Величина E(m) показывает, какая доля информации теряется при замене исходных признаковых описаний длины n на более короткие описания длины m. Метод главных компонент особенно эффективен в тех случаях, когда E(m) оказывается малым уже при малых значениях m. Для определения m обычно используют критерий «крутого обрыва». На графике E(m) отмечается то значение m, при котором происходит резкий скачок: E(m − 1) ≫ E(m), при условии, что E(m) уже достаточно мало. У нас $\lambda_i$ хорошо аппроксимируюся линейной зависимостью. График E(m) будет очень плавен, бехз резких перепадов. Поэтому перевести датасет с помощью PCA в пространство меньшей размерности с минимальными потерями точности не получится.

```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''
from scipy import diag

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n,l) = F.shape
    Sigma = 1.0 / l * np.dot(F, np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    V, d, U = np.linalg.svd((Sigma), full_matrices=False)
    #U, d, V = посчитать SVD матрицы Sigma
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / dTot * 100.0 for i in range(n)])
    # m = вычислите необходимое число главных компонент
    m = n 
    for i in range(n):
        if(var_i[i] >= varRetained * 100):
            m = i + 1
            break
    acc = (np.sum(d[:m]) / dTot * 100.0)
    print '%.2f %% variance retained in %d dimensions' % (acc, m)
    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    index = np.argsort(-d)
    select_vec = np.matrix(V.T[index[:m]])
    # compute the reduced dimensional features by projection
    #U_reduced = только m главных компонент
    U_reduced = select_vec.T 
    #G = вычислить матрицу в преобразованном пространстве
    G = np.dot(F.T, U_reduced)
    return G, U_reduced
```

### Кластеризация цифр с помощью dbscan

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_digits
from sklearn import metrics
from sklearn.metrics.cluster import adjusted_mutual_info_score, adjusted_rand_score
%matplotlib inline

digits = load_digits()
X = digits.data 
y = digits.target
```

```python
def give_score(eps, min_samples):
    db = DBSCAN(eps, min_samples)
    preds = db.fit_predict(X)
    return (adjusted_mutual_info_score(y, preds), adjusted_rand_score(y, preds))

epsilons = np.arange(1, 30)
scores = []

for epsilon in epsilons:
    scores.append(give_score(epsilon, 15))
scores = np.array(scores).T

plt.plot(epsilons, scores[0], label='AMIS')
plt.plot(epsilons, scores[1], label='ARS')
plt.xlabel('epsilon')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

График был протестирован при различных min_samples. Оптимальный вариант для epsilon - 23

```python
min_samples = np.arange(1, 30)
scores = []

for min_sample in min_samples:
    scores.append(give_score(23, min_sample))
scores = np.array(scores).T

plt.plot(min_samples, scores[0], label='AMIS')
plt.plot(min_samples, scores[1], label='ARS')
plt.xlabel('min_samples')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

Также тестировал при различных epsilon. Идеальный (на мой взгляд) вариант для min_sample - 15

```python
db = DBSCAN(eps = 23, min_samples=15).fit(X)
print("Adjusted mutual score %0.3f" % metrics.adjusted_mutual_info_score(y, db.labels_))
print("Adjusted rand score %0.3f" % metrics.adjusted_rand_score(y, db.labels_))

core_points = db.core_sample_indices_
noise_points = np.where(db.labels_ == -1)[0]
```

### Уменьшение палитры изображения

```python
import scipy
image = scipy.misc.imread('two-lorikeet-birds-2293918.jpg')
plt.imshow(image)
plt.show()
```

```python
shapes = image.shape
print shapes
image = image.reshape((shapes[0] * shapes[1], shapes[2]))
n_clusters = [2, 3, 10, 20]
```

```python
print image.shape
```

### Kmeans

```python
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering
from skimage.measure import compare_ssim

kmeans_ssim = []
for k in n_clusters:
    model = KMeans(n_clusters = k)
    model.fit(image)
    result = np.array([model.cluster_centers_[model.labels_[i]] for i in range(len(image))], dtype='uint8')
    kmeans_ssim.append(compare_ssim(image, result, multichannel=True))
    print 'n_clusters = ', str(k)
    plt.imshow(result.reshape(shapes))
    plt.show()
```

### Mean-Shift

```python
from sklearn.cluster import estimate_bandwidth
meanshift_ssim = []
quantile_list = [0.2, 0.17, 0.05, 0.03]
for q in quantile_list:
    bandwidth = estimate_bandwidth(image, quantile=q, n_samples=500)
    model = MeanShift(bandwidth=bandwidth, bin_seeding=True, n_jobs=-1)
    model.fit(image)
    result = np.array([model.cluster_centers_[model.labels_[i]] for i in range(len(image))], dtype=image.dtype)
    meanshift_ssim.append(compare_ssim(image, result, multichannel=True))
    print 'n_clusters = ' + str(len(np.unique(model.labels_)))
    plt.imshow(result.reshape(shapes))
    plt.show()
```

### AgglomerativeClustering

```python
from sklearn.neighbors import kneighbors_graph
agg_ssim = []
connectivity = kneighbors_graph(image, n_neighbors= 30, include_self=False)
for k in n_clusters:
    model = AgglomerativeClustering(n_clusters = k, connectivity=connectivity)
    model.fit(image)
    cluster_centers = [np.mean(image[model.labels_ == i], axis=0) for i in range(k)]
    result = np.array([cluster_centers[model.labels_[i]] for i in range(len(image))], dtype=image.dtype)
    agg_ssim.append(compare_ssim(image, result, multichannel=True))
    print 'n_clusters = ' + str(k)
    plt.imshow(result.reshape(shapes))
    plt.show()
```

```python
for i in range(4):
    print 'KMeans, n=', str(n_clusters[i]), 'ssim=', kmeans_ssim[i]
    print 'Mean-Shift, n=', str(n_clusters[i]), 'ssim=', meanshift_ssim[i]
    print 'AgglomerativeClustering, n=', str(n_clusters[i]), 'ssim=', agg_ssim[i]
```

#### KMeans показал наилучшее качество
