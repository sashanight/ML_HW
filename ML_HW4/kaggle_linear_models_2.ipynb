```python
i = 0
train_set = []
for line in train_data:
    if i != 0:
        b = line.split(',')
        word = b[1]
        infinitive = b[2].split('+')[0]
        word_form = b[2].split('+')[1][:-1]
        train_set.append((word, infinitive, word_form))
    i += 1
    
```
